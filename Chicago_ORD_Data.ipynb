## Project Description



### Import Libraries

```python
import numpy as np
import pandas as pd
import glob
import os
```

### Data ingestion

```python
# Ingesting flight data for Chicago(ORD)
# Departures
# Airlines: American Airlines, Delta Airlines, Southwest Airlines, and United Airlines
# 2022 - 2026
ORD_flights_files = sorted(glob.glob("Data/ORD_Dep_*_Flight_Stats.csv"))
print(f"Found {len(ORD_flights_files)} ORD Arrival file(s):")
for f in ORD_flights_files:
    print(" ", f)

ORD_flights = pd.concat(
    (pd.read_csv(f, skiprows=7) for f in ORD_flights_files), ignore_index=True)
print(f"Total flight rows loaded: {len(ORD_flights)}")
```

```python
# ingesting chicago weather data 
# 01/01/2022 - 07/09/2026
chi_weather_files = sorted(glob.glob("Data/ORD_Weather_*.psv") + glob.glob("Data/ORD_Weather_2022"))
print(f"Found {len(chi_weather_files)} Chicago weather file(s):")
for f in chi_weather_files:
    print(" ", f)

chi_weather = pd.concat(
    (pd.read_csv(f, sep="|") for f in chi_weather_files), ignore_index=True)
print(f"Total weather rows loaded: {len(chi_weather)}")
```

```python
# ingesting aircraft data
aircraft = pd.read_csv("Data/MASTER.txt", sep=",", encoding="utf-8-sig", low_memory=False)
print(f"Aircraft registry rows loaded: {len(aircraft)}")
```
