### Merge Departures/FAA on Tail Number

```python
# Normalizing tail numbers in both datasets to create a common key for merging
departures_df["tail_key"] = (
    departures_df["mod_tail_number"]
    .astype("string")
    .str.strip()
    .str.upper()
    .str.replace(r"^N", "", regex=True)
)

faa_df["tail_key"] = (
    faa_df["N-NUMBER"]
    .astype("string")
    .str.strip()
    .str.upper()
    .str.replace(r"^N", "", regex=True)
)

departures_faa_df = departures_df.merge(
    faa_df,
    on="tail_key",
    how="left",
    validate="many_to_one",
    indicator=True,
)

print(f"departures_faa_df shape: {departures_faa_df.shape}")
display(departures_faa_df.head())
```

### Merge Departures/FAA and NOAA Data using time horizons
Using the time horizons (1, 2, 4, 6, 12) hours before departure, we will end up with five merged datasets. For simplicity, we will start by merging weather data 1 hour before departure and write a function to create the other datasets.

```python
# Only use weather feature columns
weather_features = [
    "visibility",
    "ceiling_height",
    "wind_speed",
    "wind_direction",
    "precipitation",
    "temperature",
    "dew_point_temperature",
    "relative_humidity",
    "altimeter",
    "sky_cover_summation_baseht_1"
]

noaa_df = (
    noaa_df[["DATE"] + weather_features]
    .rename(columns={"DATE": "weather_source_dt"})
)
```

```python
departures_faa_cleaned["scheduled_dep_dt"] = pd.to_datetime(
    departures_faa_cleaned["scheduled_dep_dt"], errors="coerce"
)
departures_faa_cleaned["weather_cutoff_dt"] = (
    departures_faa_cleaned["scheduled_dep_dt"] - pd.Timedelta(hours=1)
)

noaa_df["weather_observation_dt"] = pd.to_datetime(
    noaa_df["weather_source_dt"], errors="coerce"
)

departures_faa_noaa_df = pd.merge_asof(
    departures_faa_cleaned.dropna(subset=["weather_cutoff_dt"]).sort_values("weather_cutoff_dt"),
    noaa_df.dropna(subset=["weather_observation_dt"]).sort_values("weather_observation_dt"),
    left_on="weather_cutoff_dt",
    right_on="weather_observation_dt",
    direction="backward",
    tolerance=pd.Timedelta(minutes=60),
    suffixes=("", "_weather"),
)

print(f"Merged shape: {departures_faa_noaa_df.shape}")
print(f"Weather match rate: {departures_faa_noaa_df['weather_observation_dt'].notna().mean():.2%}")
display(departures_faa_noaa_df.head())
```

### Additional engineered fields

**Aircraft and calendar**

- `aircraft_age`: flight year minus aircraft manufacture year
- `week_num`: 1-based seven-day period beginning January 1

**Cyclical time**

- `sched_dep_hour_sin`, `sched_dep_hour_cos`: circular representation of scheduled departure hour (PENDING)
- `month_sin`, `month_cos`: circular representation of month (PENDING)

**Weather**

- `wind_direction_sin`, `wind_direction_cos`: circular representation of wind direction (PENDING)
- `temperature_dewpoint_spread`: temperature minus dew point

**Recent airport conditions**

- `airport_delay_average_1h`: average delay observed in the hour before the prediction cutoff
- `airport_delay_stddev_1h`: sample delay stddev over that window
- `airport_departures_observed_1h`: number of observed departures supporting those statistics

The prediction cutoff is one hour before scheduled departure. Only flights that had departed by the cutoff are used, and the current flight is excluded to prevent data leakage.

#### Aircraft and calendar fields

```python
import numpy as np
from pandas.tseries.holiday import USFederalHolidayCalendar

departures_faa_noaa_cleaned = departures_faa_noaa_cleaned.copy()
departures_faa_noaa_cleaned["date"] = pd.to_datetime(departures_faa_noaa_cleaned["date"], errors="coerce")
departures_faa_noaa_cleaned["scheduled_dep_dt"] = pd.to_datetime(departures_faa_noaa_cleaned["scheduled_dep_dt"], errors="coerce")
departures_faa_noaa_cleaned["year_mfr"] = pd.to_numeric(departures_faa_noaa_cleaned["year_mfr"], errors="coerce")

# Aircraft age
departures_faa_noaa_cleaned["aircraft_age"] = departures_faa_noaa_cleaned["year"] - departures_faa_noaa_cleaned["year_mfr"]
departures_faa_noaa_cleaned.loc[departures_faa_noaa_cleaned["aircraft_age"] < 0, "aircraft_age"] = np.nan

# Seven-day periods relative to January 1
departures_faa_noaa_cleaned["week_num"] = ((departures_faa_noaa_cleaned["date"].dt.dayofyear - 1) // 7 + 1).astype("int8")
```

#### Recent airport delay fields

To derive the recent airport conditions, we define a lookback window of 1 hour. Only using delays from flights whos delays are already observable in the lookback window.

e.g. 
scheduled_departure = 3:00 PM
prediction time     = 2:00 PM (time horizon of 1h)
lookback window     = 1:00 PM to 2:00 PM

```python
HORIZON_HOURS = 1
LOOKBACK_HOURS = 1

airport_events = departures_df[["actual_dep_dt", "departure_delay_minutes"]].copy()
airport_events["actual_dep_dt"] = pd.to_datetime(airport_events["actual_dep_dt"], errors="coerce")
airport_events["departure_delay_minutes"] = pd.to_numeric(airport_events["departure_delay_minutes"], errors="coerce")
airport_events = airport_events.dropna(subset=["actual_dep_dt", "departure_delay_minutes"]).sort_values("actual_dep_dt")
```

### Flight lookback helper function

```python
def get_recent_airport_delay_stats(row):
    scheduled_departure = row["scheduled_dep_dt"]

    prediction_cutoff = scheduled_departure - pd.Timedelta(hours=HORIZON_HOURS)
    window_start = prediction_cutoff - pd.Timedelta(hours=LOOKBACK_HOURS)

    recent_departures = airport_events[
        (airport_events["actual_dep_dt"] >= window_start)
        & (airport_events["actual_dep_dt"] <= prediction_cutoff)
    ]

    return pd.Series({
        "airport_delay_average_1h": recent_departures["departure_delay_minutes"].mean(),
        "airport_delay_stddev_1h": recent_departures["departure_delay_minutes"].std(),
        "airport_departures_observed_1h": len(recent_departures)
    })
```

```python
departures_faa_noaa_cleaned["scheduled_dep_dt"] = pd.to_datetime(departures_faa_noaa_cleaned["scheduled_dep_dt"], errors="coerce")

recent_delay_features = departures_faa_noaa_cleaned.apply(get_recent_airport_delay_stats, axis=1)

df = pd.concat([departures_faa_noaa_cleaned, recent_delay_features], axis=1)
```

#### Airport delay average and stddev

- `airport_delay_average_1h` is null when **no qualifying departures** occurred during the one-hour lookback window. With zero delay values, there is no average to calculate.
- `airport_delay_stddev_1h` is null when the window contains **fewer than two qualifying departures**. Pandas calculates sample stddev with `ddof=1`, which requires at least two values.

Hence, we will fillna(0)

```python
departures_faa_noaa_cleaned["airport_delay_average_1h"] = (departures_faa_noaa_cleaned["airport_delay_average_1h"].fillna(0.0))
departures_faa_noaa_cleaned["airport_delay_stddev_1h"] = (departures_faa_noaa_cleaned["airport_delay_stddev_1h"].fillna(0.0))

# Keep the target as the final column after adding the two history flags.
target_column = "departure_delay_minutes"
departures_faa_noaa_cleaned = departures_faa_noaa_cleaned[[
    column for column in departures_faa_noaa_cleaned.columns
    if column != target_column
] + [target_column]]
```
