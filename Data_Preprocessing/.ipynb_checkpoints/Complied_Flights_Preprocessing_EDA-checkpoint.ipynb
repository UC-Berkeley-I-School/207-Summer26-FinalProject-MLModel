### Import Libraries

```python
import numpy as np
import pandas as pd
import glob
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns

%matplotlib inline
```

### Data ingestion

```python
# Ingesting flight data for SFO
# Departures
# Airlines: Alaskan Airlines, American Airlines, Delta Airlines, Frontier Airlines
# 2022 - 2026
SFO_flights_files = sorted(glob.glob("Data_Extraction/Data/SFO_*_Departures.csv"))
for f in SFO_flights_files:
    print(" ", f)

SFO_flights = pd.concat(
    (pd.read_csv(f, skiprows=7) for f in SFO_flights_files), ignore_index=True)
print(f"Total flight rows loaded: {len(SFO_flights)}")
```

### Data Preprocessing


```python
#shape and dtypes for flights
print("Shape of flights df:", SFO_flights.shape)
print("Data types:")
print(SFO_flights.dtypes)
SFO_flights.head()
```

```python
SFO_flights.to_csv("SFO_flights_combined.csv", index=False)
```

```python
raw_flights = pd.read_csv('Data_extraction/SFO_flights_combined.csv')
```

```python
#Creating the a copy of the dataset.
flights = raw_flights.copy()
```

```python
#Standardizing column names
flights.columns = (
    flights.columns
    .str.strip()
    .str.lower()
    .str.replace(r'[\s\(\)/\-]+', '_', regex=True)
    .str.replace(r'[^\w]', '', regex=True)
    .str.strip('_')
)

print(flights.columns.tolist())
flights.head()
```

```python
# Create cancelled indicator
flights['is_cancelled'] = (pd.to_numeric(flights['actual_elapsed_time_minutes'], errors='coerce') == 0).astype(int)

# Convert delay column to numeric once
delay = pd.to_numeric(flights['departure_delay_minutes'], errors='coerce')

# Assign outcome classes
flights['outcome'] = np.select(
    [
        flights['is_cancelled'] == 1,        
        delay <= 0,                          
        delay.between(1, 59),                
        delay.between(60, 119),              
        delay.between(120, 179),             
        delay.between(180, 239),             
        delay.between(240, 299),             
        delay.between(300, 359),             
        delay >= 360],
    [   8,
        0,
        1,
        2,
        3,
        4,
        5,
        6,
        7],
    default=np.nan)

# Regression target
flights['departure_delay_min'] = np.where(
    flights['is_cancelled'] == 1,
    np.nan,
    delay)

labels = {
    0: 'No delay or early',
    1: '<1 hour delay',
    2: '1-2 hour delay',
    3: '2-3 hour delay',
    4: '3-4 hour delay',
    5: '4-5 hour delay',
    6: '5-6 hour delay',
    7: '>6 hour delay',
    8: 'Cancelled'}

counts = flights['outcome'].value_counts().sort_index()

print("Outcome distribution:")
for k, v in counts.items():
    print(f"{k}: {labels[k]}: {v:,}")
```

```python
#Combining date and time to a proper datetime field
flights['scheduled_dep_dt'] = pd.to_datetime(
   flights['date_mm_dd_yyyy'].astype(str) + ' ' + flights['scheduled_departure_time'].astype(str),
    format='%m/%d/%Y %H:%M',
    errors='coerce'
)

flights['actual_dep_dt'] = pd.to_datetime(
  flights['date_mm_dd_yyyy'].astype(str) + ' ' + flights['actual_departure_time'].astype(str),
    format='%m/%d/%Y %H:%M',
    errors='coerce'
)

flights['date'] = pd.to_datetime(flights['date_mm_dd_yyyy'], format='%m/%d/%Y', errors='coerce')

#Only using data available up to 6 hours before the flight is scheduled to depart.
flights['6hr_cutoff_dt'] = flights['scheduled_dep_dt'] - pd.Timedelta(hours=6)

flights[['date', 'scheduled_dep_dt', 'actual_dep_dt', '6hr_cutoff_dt']].head()
```

```python
# Convert useful columns to appropriate types.
flights['tail_number'] = flights['tail_number'].astype('string')
flights['destination_airport'] = flights['destination_airport'].astype('string')
flights['carrier_code'] = flights['carrier_code'].astype('string')

# Numeric columns from BTS data.
numeric_cols = [
    'scheduled_elapsed_time_minutes', 'actual_elapsed_time_minutes',
    'departure_delay_minutes', 'taxiout_time_minutes',
    'delay_carrier_minutes', 'delay_weather_minutes',
    'delay_national_aviation_system_minutes', 'delay_security_minutes',
    'delay_late_aircraft_arrival_minutes'
]

for col in numeric_cols:
    if col in flights.columns:
       flights[col] = pd.to_numeric(flights[col], errors='coerce')

print('The data type of the flights data frame after preprocessing:')
print(flights.dtypes)
flights.head()
```

```python
# Missing value summary.
missing = flights.isna().sum()
missing_percent = (missing / len(flights) * 100) 
summary = (
    pd.DataFrame({'missing_count': missing, 'missing_percent': missing_percent})
    .query('missing_count > 0')
    .sort_values('missing_percent', ascending=False)
)
print(f'Columns with missing values ({len(summary)} of {len(flights.columns)}):')
summary
```

```python
# Cause of delay counts.
cause_cols = [
    'delay_carrier_minutes',
    'delay_weather_minutes',
    'delay_national_aviation_system_minutes',
    'delay_security_minutes',
    'delay_late_aircraft_arrival_minutes'
]

op_all = flights[flights['is_cancelled'] == 0]
for c in cause_cols:
    if c in op_all.columns:
        n = (op_all[c] > 0).sum()
        total = op_all[c].sum()
        print(f'  {c}: {n:,} flights, {total:,.0f} total minutes')
```

```python
# Relevant datetime features.
flights['year'] = flights['date'].dt.year
flights['month'] = flights['date'].dt.month
flights['month_name'] = flights['date'].dt.strftime('%b')
flights['day_of_week'] = flights['date'].dt.dayofweek
flights['is_weekend'] = (flights['day_of_week'] >= 5).astype(int)
flights['sched_dep_hour'] = flights['scheduled_dep_dt'].dt.hour

flights.head()
```

```python
# Time of day column.
flights['dep_time_of_day'] = pd.cut(
   flights['sched_dep_hour'],
    bins=[0, 6, 12, 18, 24],
    labels=['early morning', 'morning', 'afternoon', 'evening/night'],
    right=False
)

flights.head()
```

```python
#holiday flags
us_holidays = pd.to_datetime([
    '2022-01-01','2023-01-01','2024-01-01','2025-01-01','2026-01-01',
    '2022-05-30','2023-05-29','2024-05-27','2025-05-26','2026-05-25',
    '2022-07-04','2023-07-04','2024-07-04','2025-07-04','2026-07-04',
    '2022-09-05','2023-09-04','2024-09-02','2025-09-01','2026-09-07',
    '2022-11-24','2023-11-23','2024-11-28','2025-11-27','2026-11-26',
    '2022-12-25','2023-12-25','2024-12-25','2025-12-25','2026-12-25',
])
flights['is_holiday'] = flights['date'].isin(us_holidays).astype(int)
print(f'Flights on holidays: {flights["is_holiday"].sum():,}')
```

```python
# Holiday distance features

# Make sure date is datetime
flights["date"] = pd.to_datetime(flights["date"])

# Your existing US holiday list should already exist as us_holidays
holiday_dates = pd.to_datetime(list(us_holidays)).sort_values()

def days_until_next_holiday(date):
    future_holidays = holiday_dates[holiday_dates >= date]
    if len(future_holidays) == 0:
        return np.nan
    return (future_holidays[0] - date).days

def days_from_nearest_holiday(date):
    return np.min(np.abs((holiday_dates - date).days))

flights["days_until_holiday"] = flights["date"].apply(days_until_next_holiday)
flights["days_from_holiday"] = flights["date"].apply(days_from_nearest_holiday)

flights[["date", "is_holiday", "days_until_holiday", "days_from_holiday"]].head()
```

```python
# Strip the leading 'N' in the tail number so it matches with the aircraft dataset.
flights['mod_tail_number'] = flights['tail_number'].str.strip().str.lstrip('N')

flights['carrier_code'] = flights['carrier_code'].astype('category')
flights['destination_airport'] = flights['destination_airport'].astype('category')
flights['dep_time_of_day'] = flights['dep_time_of_day'].astype('category')

print(f'Unique carriers: {flights["carrier_code"].nunique()}')
print(f'Unique destinations: {flights["destination_airport"].nunique()}')
flights.head()
```

```python
flights.to_csv("SFO_flights_preprocessed.csv", index=False)
```

### Data Splits

```python
# Sort the data by date/scheduled departure datetime before splitting.
flights = flights.sort_values(['date', 'scheduled_dep_dt']).reset_index(drop=True)

# Calculate the number of samples based on the split percentage.
total_rows = len(flights)
train_end_idx = int(total_rows * 0.60)
val_end_idx = train_end_idx + int(total_rows * 0.20)

# Split the data in chronological order.
train = flights.iloc[:train_end_idx].copy()
val = flights.iloc[train_end_idx:val_end_idx].copy()
test = flights.iloc[val_end_idx:].copy()

print(f'Train shape: {train.shape}')
print(f'Validation shape: {val.shape}')
print(f'Test shape: {test.shape}')
```

```python
print(f"Train: {train['date'].min()} to {train['date'].max()}")
print(f"Validation: {val['date'].min()} to {val['date'].max()}")
print(f"Test: {test['date'].min()} to {test['date'].max()}")
```

```python
# From this point forward, EDA will only use training set.
op = train[train['is_cancelled'] == 0].copy()
print(f'Operational/non-cancelled training flights: {len(op):,}')
```

### Exploratory Data Analysis (EDA)

```python
# Target distribution in the training set.
labels = {
    0: 'No delay or early',
    1: '<1 hour delay',
    2: '1-2 hour delay',
    3: '2-3 hour delay',
    4: '3-4 hour delay',
    5: '4-5 hour delay',
    6: '5-6 hour delay',
    7: '>6 hour delay',
    8: 'Cancelled'}
train_counts = train['outcome'].map(labels).value_counts()

plt.figure(figsize=(8, 5))
sns.countplot(data=train, x='outcome', color='steelblue', legend=False)
plt.title('Training Set Outcome Distribution', fontsize=12, pad=20)
plt.xlabel('Outcome Class')
plt.ylabel('Flight Count')
plt.xticks([0, 1, 2, 3, 4, 5, 6, 7, 8],[labels[0], labels[1], labels[2], labels[3], labels[4],labels[5], labels[6], labels[7], labels[8]],
    rotation=30,
    ha='right')
plt.tight_layout()
plt.show()

train_counts
```

```python
airline_counts = (train['carrier_code']
    .value_counts()
    .sort_values())
fig, ax = plt.subplots(figsize=(9,5))
bars = ax.barh(airline_counts.index, airline_counts.values, color='steelblue')
ax.set_title('Number of Flights by Airline (Training Set)', fontsize=15, fontweight='bold')
ax.set_xlabel('Number of Flights')
ax.set_ylabel('Airline')
for bar in bars:
    ax.text(
        bar.get_width() + 1500,
        bar.get_y() + bar.get_height()/2,
        f'{int(bar.get_width()):,}',
        va='center',
        fontsize=9)
plt.grid(axis='x', alpha=0.25)
plt.tight_layout()
plt.show()
```

```python
top_destinations = (train['destination_airport']
                    .value_counts() 
                    .head(20) 
                    .sort_values())

plt.figure(figsize=(10,7))
top_destinations.plot(kind='barh', color='steelblue', edgecolor='white')
plt.title('Top 20 Destination Airports by Flight Count (Training Set)', fontsize=14, fontweight='bold')
plt.xlabel('Number of Flights')
plt.ylabel('Destination Airport')
plt.tight_layout()
plt.show()
```

```python
# Monthly flight trends using training data only.
monthly_counts = (
    train.groupby(['year', 'month', 'month_name'])
    .size()
    .reset_index(name='flight_count')
)

# Keep months in calendar order.
monthly_counts['month_name'] = pd.Categorical(
    monthly_counts['month_name'],
    categories=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'],
    ordered=True
)

# Create the chart.
g = sns.relplot(
    data=monthly_counts,
    x='month_name',
    y='flight_count',
    col='year',
    col_wrap=2,
    kind='line',
    marker='o',
    linewidth=2.5,
    color='teal',
    height=4,
    aspect=1.5,
    facet_kws=dict(sharex=False),
)

g.set_titles('{col_name} Monthly Flight Trends')
g.set_axis_labels('Month', 'Flights Count')
for ax in g.axes.flat:
    ax.tick_params(axis='x', labelbottom=True, labelrotation=45)
    ax.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

```python
# Create the correlation matrix for delay-related numeric columns.
timedelta_cols = [
    'scheduled_elapsed_time_minutes', 'actual_elapsed_time_minutes',
    'departure_delay_minutes', 'taxiout_time_minutes',
    'delay_carrier_minutes', 'delay_weather_minutes',
    'delay_national_aviation_system_minutes', 'delay_security_minutes',
    'delay_late_aircraft_arrival_minutes'
]
timedelta_cols = [c for c in timedelta_cols if c in train.columns]

corr_data = train[timedelta_cols].apply(pd.to_numeric, errors='coerce').fillna(0)
correlation_matrix = corr_data.corr()
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

plt.figure(figsize=(12, 10))
sns.heatmap(
    correlation_matrix,
    mask=mask,
    annot=True,
    cmap='mako',
    fmt='.2f',
    linewidths=0.5,
    cbar_kws={'shrink': 0.8}
)
plt.title('Delay-related Correlation Matrix - Training Set')
plt.show()
```

```python
# Total delay minutes by cause.
cause_cols = {
    'delay_carrier_minutes': 'Carrier',
    'delay_weather_minutes': 'Weather',
    'delay_national_aviation_system_minutes': 'NAS',
    'delay_security_minutes': 'Security',
    'delay_late_aircraft_arrival_minutes': 'Late Aircraft'
}
cause_cols = {k: v for k, v in cause_cols.items() if k in op.columns}

cause_totals = op[list(cause_cols.keys())].sum().rename(cause_cols)

fig, ax = plt.subplots(figsize=(10, 4))
cause_totals.sort_values().plot(kind='barh', ax=ax, color='steelblue', edgecolor='white')
ax.set_title('Total Delay Minutes by Cause - Training Set', fontsize=12)
ax.set_xlabel('Total Delay (minutes)')
plt.tight_layout()
plt.show()
```

```python
# Distribution of departure delays by carrier.
carriers = sorted(op['carrier_code'].dropna().unique())

fig, axes = plt.subplots(2, 4, figsize=(16, 8), sharex=True, sharey=True)
axes = axes.flatten()

for ax, carrier in zip(axes, carriers):
    data = op.loc[op['carrier_code'] == carrier, 'departure_delay_min'].clip(-30, 150)
    ax.hist(
        data,
        bins=60,
        color='steelblue',
        edgecolor='white',
        linewidth=0.3)
    ax.axvline(15, color='firebrick', linestyle='--', linewidth=1)
    ax.set_title(f"{carrier}", fontsize=11, fontweight='bold')
    ax.set_xlabel("Delay (min)", fontsize=9)
    ax.set_ylabel("Flights", fontsize=9)

fig.suptitle(
    "Distribution of Departure Delays by Airline (Training Set)",
    fontsize=16,
    fontweight='bold')
plt.tight_layout(rect=[0,0,1,0.95])
plt.show()
```

```python
# Distribution of delay by month and day of week.
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

month_delay = op.groupby('month')['departure_delay_min'].mean()
month_labels = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
axes[0].bar(month_delay.index, month_delay.values, color='steelblue', edgecolor='white')
axes[0].set_xticks(range(1, 13))
axes[0].set_xticklabels(month_labels)
axes[0].set_title('Mean Departure Delay by Month - Training Set', fontsize=12)
axes[0].set_ylabel('Mean Delay (minutes)')

dow_delay = op.groupby('day_of_week')['departure_delay_min'].mean()
dow_labels = ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']
axes[1].bar(dow_delay.index, dow_delay.values, color='steelblue', edgecolor='white')
axes[1].set_xticks(range(7))
axes[1].set_xticklabels(dow_labels)
axes[1].set_title('Mean Departure Delay by Day of Week - Training Set', fontsize=12)
axes[1].set_ylabel('Mean Delay (minutes)')

plt.tight_layout()
plt.show()
```

```python
# Get the flight counts based on tail number.
flight_counts = train['tail_number'].value_counts()

# Calculate the average delay by tail number then only select the top 10.
top_delayed_tail_num = (
    train.groupby('tail_number')['delay_late_aircraft_arrival_minutes']
    .mean()
    .nlargest(10)
)

plt.figure(figsize=(10, 6))
top_delayed_tail_num.plot(kind='bar', color='salmon', edgecolor='black')
plt.title('Top 10 Airplane Tail Numbers with the Worst Average Delays - Training Set', fontsize=14, pad=15)
plt.ylabel('Average Late Aircraft Arrival Delay (Minutes)', fontsize=12)
plt.xlabel('Tail Number', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Calculate average delay for each destination and select the top 30 to include in the chart.
top_delayed_airports = (
    train.groupby('destination_airport')['delay_late_aircraft_arrival_minutes']
    .mean()
    .nlargest(30)
)

plt.figure(figsize=(12, 8))
top_delayed_airports.plot(kind='barh', color='skyblue', edgecolor='black')
plt.title('Top Destination Airports with the Highest Average Departure Delays - Training Set', fontsize=14, pad=15)
plt.xlabel('Average Late Aircraft Arrival Delay (Minutes)', fontsize=12)
plt.ylabel('Destination Airport', fontsize=12)
plt.gca().invert_yaxis()
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
