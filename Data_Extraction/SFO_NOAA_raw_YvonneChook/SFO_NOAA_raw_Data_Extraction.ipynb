```python
#Load Libraries
import os
from pathlib import Path
import pandas as pd
import gdown
```

```python
def load_sfo_noaa_data():
    # Google Drive folder with SFO NOAA PSV files
    url = "https://drive.google.com/drive/folders/1qZHc7AxcL_1GaD2ZTN2n8O7lP7VmoLoh"
    folder = "data"
    output = "NOAA_SFO_2022_2026.parquet"
    
    # Download only if the folder doesn't exist
    if not os.path.exists(folder):
        gdown.download_folder(url, output=folder, quiet=False, use_cookies=False)
    
    # Load all .psv files
    psv_files = list(Path(folder).rglob("*.psv"))
    
    dfs = []
    for file in psv_files:
        df = pd.read_csv(file, sep="|", dtype=str, low_memory=False)
        dfs.append(df)
    
    # Combine and export
    noaa_sfo_df = pd.concat(dfs, ignore_index=True)
    noaa_sfo_df.to_parquet(output, index=False)
    
    return noaa_sfo_df

NOAA_SFO_df = load_sfo_noaa_data()
```

```python
# Analyze data downloaded
print("NOAA SFO 2022-2026:")
print(f"\nShape: {NOAA_SFO_df.shape}")

print("\nColumns and Datatypes:")
display(NOAA_SFO_df.dtypes)

print("\nData Preview:")
display(NOAA_SFO_df.head())
```
