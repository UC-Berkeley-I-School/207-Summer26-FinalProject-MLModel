```python
#Upload Libraries

import numpy as np
import pandas as pd
import glob
```

```python
# Arrivals Data from US DOT by Site

# Get only real CSV files, not Zone.Identifier files
USDOT = [
    file for file in glob.glob("data/Arrivals_data_by_airport/*.csv")
    if not file.endswith(":Zone.Identifier")
]

#verfiy number of files found
print("Files found:", len(USDOT))
print(USDOT)

#skip the first 7 rows of non-data
df = pd.concat(
    [pd.read_csv(file, skiprows=7, skipfooter=1, engine="python") for file in USDOT],
    ignore_index=True
)

print(df.shape)
df.head()

#export to .csv file
#df.to_csv("DEN_arrivals.csv", index=False)
```

```python
# Weather info in DEN International airport

# NOAA weather files
weather_files = glob.glob("data/weather_info_by_year/*.psv")

print("Files found:", len(weather_files))
print(weather_files)

# Load and append all weather files
weather_df = pd.concat(
    [pd.read_csv(file, sep="|") for file in weather_files],
    ignore_index=True
)

print(weather_df.shape)
weather_df.head()

#export to .csv file
#df.to_csv("DEN_weather.csv", index=False)
```

```python
#aircraft information

aircraft_master = pd.read_csv(
    "data/Aircraft Information/MASTER.txt",
    dtype=str
)

print(aircraft_master.shape)
aircraft_master.head()

#export to .csv file
#df.to_csv("aircraft_info.csv", index=False)
```
